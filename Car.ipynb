# Read the dataset to the python environment.

```python
import pandas as pd
car = pd.read_csv('cars_data.csv')
car
```

# Check for the null values present in the dataset.

```python
car.isnull().sum()
```

# Plot a bar graph of male vs female buyers participated in the sales. 

```python
car['Buyer Gender'].value_counts()
```

```python
import matplotlib.pyplot as plt
car['Buyer Gender'].value_counts().plot(kind='bar',color=['blue','red'])
plt.xlabel("Gender", labelpad=14)
plt.ylabel("Count of People", labelpad=14)
plt.title(" male vs female buyers participated in the sales", y=1.02);
```

# Find the top 5 cars based on their sales price.

```python
df = car.nlargest(5,['Sale Price'])
df[['Make','Model','Sale Price']]
```

# Find the least 5 cars based on their Resell price.

```python
df = car.nsmallest(5,['Resell Price'])
df[['Make','Model','Resell Price']]
```
